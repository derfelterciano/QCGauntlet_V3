```python
import pandas as pd, numpy as np,

dataset_1_ds = "/Users/dterciano/Desktop/LokeyLabFiles/TargetMol/Datasets/1umData/TargetMol_PMA_HistDiff_concatanated_UPDATED.csv"
dataset_2_ds = "/Users/dterciano/Desktop/LokeyLabFiles/TargetMol/Datasets/1umData/TargetMol_DMSO_HistDiff_concatanated_UPDATED.csv"

dataset_1_ds = pd.read_csv(dataset_1_ds, index_col=0)
dataset_2_ds = pd.read_csv(dataset_2_ds, index_col=0)

annotation_sheet = '/Users/dterciano/Desktop/LokeyLabFiles/TargetMol/Annotations/L4000-Bioactive Compound Library-Beverley Rabbitts (KIT10006535)_MapCleaned3_KSready.csv'
annotation_sheet = pd.read_csv(annotation_sheet, index_col=1)

display(dataset_1_ds, dataset_2_ds, annotation_sheet)
```

```python
def match_annotations(ds: pd.DataFrame, annots: pd.DataFrame) -> pd.DataFrame:
    merge_df = pd.merge(left=ds, right=annotation_sheet[['longname_proper']], left_index=True, right_index=True, how='left')
    merge_df.index.name = "index"
    merge_df.reset_index(inplace=True)
    merge_df[['well', 'plate']] = merge_df['index'].str.split('._.', regex=False, expand=True)

    merge_df['longname_proper'] = merge_df['longname_proper'].apply(lambda x: x.split('._.')[0])
    merge_df.drop('index', axis=1, inplace=True)
    
    return merge_df.set_index(['longname_proper', 'well', 'plate'])


df = match_annotations(ds=dataset_1_ds, annots=annotation_sheet)

df  
```

```python
import os
```

```python
dataset_1_ds = match_annotations(ds=dataset_1_ds, annots=annotation_sheet)
dataset_2_ds = match_annotations(ds=dataset_2_ds, annots=annotation_sheet)

outDir = "/Users/dterciano/Desktop/LokeyLabFiles/QCG_TestData"
dataset_1_ds.to_csv(os.path.join(outDir, 'primary_ds.csv'))
dataset_2_ds.to_csv(os.path.join(outDir, 'secondary_ds.csv'))
```
